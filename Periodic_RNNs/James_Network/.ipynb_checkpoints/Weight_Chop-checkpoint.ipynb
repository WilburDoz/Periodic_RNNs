```python
import torch
import parameters_will
import torch.optim as optim
import RNN_Will as _model_
import numpy as np
import matplotlib.pyplot as plt
import pickle
import utils

%load_ext autoreload
%autoreload 2

# Set up our parameters
params = parameters_will.default_params()
path = '/Users/will/Documents/Inverse_Models/Periodic_RNNs/James_Network/'
name = 'winner'
path_name = path + name
model = torch.load(path_name)

device = 'cpu'
```

```python
# Let's extract all the weights, biases, and the activities on a big set of frequencies
freqs = np.arange(1, 25)
jitter = 4
T = 300

# Weights and biases
out_weights = model.predict.state_dict()['weight'][0,:].detach().numpy()
out_bias = model.predict.state_dict()['bias'].detach().numpy()
in_weights = model.embedding.state_dict()['weight'][:,0].detach().numpy()
in_biases = model.embedding.state_dict()['bias'].detach().numpy()
hidden_weights = model.transition.state_dict()['weight'].detach().numpy()
hidden_biases = model.transition.state_dict()['bias'].detach().numpy()
hidden_biases = hidden_biases + in_biases # There's redundency in the model, squash it!

# Set up data
jitters = np.full(shape=len(freqs), fill_value=jitter, dtype=np.int32)
params.data.batch_size = len(freqs)
plot_batch_size = len(freqs)
model.batch_size = params.data.batch_size
params.data.min_beats = 5
params.data.min_length = T
params.data.max_length = T+1
input_dict = utils.generate_data(params, freqs, jitters)
titles = freqs


variables = model(input_dict, device=device)
output = torch.stack(variables.pred).detach().numpy()[:,:,0]
hidden = torch.stack(variables.hidden).detach().numpy()

# Great, let's zoom into one frequency and one timeperiod
cutoff_time = 150
freq_index = 4

# Cutout
out = output[cutoff_time:, freq_index]
hid = hidden[cutoff_time:, freq_index, :]

# We want to build a system to chop out the dead guys, let's begin by getting a numpy equivalent system
def ReLU(x):
    return x * (x > 0)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def run_forward(start_activity, in_weights, hidden_weights, hidden_biases, out_weights, out_bias, t = T - cutoff_time):
    h = np.zeros([t, len(start_activity)])
    o = np.zeros([t])
    h[0,:] = start_activity
    o[0] = sigmoid(np.dot(out_weights, h[0,:]) + out_bias)
    
    for tt in range(1,t):
        input_current = o[tt-1]*in_weights
        h[tt,:] = ReLU(np.matmul(hidden_weights, h[tt-1,:]) + hidden_biases + input_current)
        o[tt] = sigmoid(np.dot(out_weights, h[tt,:]) + out_bias)
        
    return (h,o)

(h,o) = run_forward(hid[0,:], in_weights, hidden_weights, hidden_biases, out_weights, out_bias, T - cutoff_time)
plt.figure(figsize = (12, 12))
plt.plot(o)

# Now let's zoom in, see what all the neurons are doing, and try to find the dead ones.
num_plots_sqrt = 10
plt.figure(figsize = (60,60))
for neuron in range(params.model.h_size):
    plt.subplot(num_plots_sqrt, num_plots_sqrt, neuron+1)
    plt.plot(h[:,neuron])
    plt.title(neuron)
    plt.xticks([])
    #plt.yticks([])
    
dead_neurons = np.sum(np.abs(h), axis = 0) < 1e-10
dead_indices = np.where(dead_neurons)[0]
print(f"{sum(dead_neurons)} dead: {dead_indices}")

print(np.linalg.norm(o - out))
print(np.linalg.norm(h - hid))
```
